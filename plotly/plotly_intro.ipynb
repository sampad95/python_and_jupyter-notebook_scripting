```python
import plotly.express as px
import numpy as np

x = np.linspace(-2, 4, 100)
x0 = 1

fig = px.line(x=x, y=0.5*(x-x0)**2, labels=dict(x='x', y='V(x)'))
fig.update_traces(line=dict(width = 3, color = 'palevioletred'))

#fig.show()
fig.show('png')
```
