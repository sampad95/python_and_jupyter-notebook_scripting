```python
def psi_n(n, a2, h):
    l=a2-0
    psi = []
    
    if type(n) == int and n != 0:
        for x in np.arange(0, a2+h, h):
            psi += [math.sqrt(2/l)*math.sin((n*math.pi*x)/l)]            
    else:
        print("Wrong input")
        
    return psi

def pd_n(n, a2, h):
    l=a2-0
    pd = []
    
    if type(n) == int and n != 0:
        for x in np.arange(0, a2+h, h):
            pd += [(math.sqrt(2/l)*math.sin((n*math.pi*x)/l))**2]            
    else:
        print("Wrong input")
        
    return pd

def xvec(a2, h):
    return np.arange(0, a2+h, h)
```

```python
def E_n(n, a2):
    l = a2-0 
    return (n**2)*(sc.h)**2/(8*sc.m_e*l**2)
```

```python
qn = 2
xmax = 1
step = 0.001

plt.vlines(x=(xmax-0)/2, ymin=-2, ymax=2, ls="--", color="grey")
plt.hlines(y=0, xmin=0, xmax=xmax, ls="--", color="grey")

plt.xlim(0, xmax)
plt.ylim(-2, 2)
plt.plot(xvec(xmax, step),  psi_n(qn, xmax, step))

plt.show()
```
