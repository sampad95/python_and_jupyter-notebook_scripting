```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define constants
L = 1  # Length of the box
n_x, n_y, n_z = 1, 2, 2  # Quantum numbers

# Define the wavefunction
def psi(x, y, z, n_x, n_y, n_z, L):
    return np.sqrt(8 / L**3) * np.sin(n_x * np.pi * x / L) * np.sin(n_y * np.pi * y / L) * np.sin(n_z * np.pi * z / L)

# Generate grid points in 3D
N = 50  # Number of points in each direction
x = np.linspace(0, L, N)
y = np.linspace(0, L, N)
z = np.linspace(0, L, N)
X, Y, Z = np.meshgrid(x, y, z)

# Compute the wavefunction at each grid point
Psi = psi(X, Y, Z, n_x, n_y, n_z, L)

# Plot the wavefunction
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot the wavefunction as a 3D surface
ax.scatter(X, Y, Z, c=Psi, cmap='viridis')

# Labels and title
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.set_title(f'Wavefunction of Particle in 3D Box (n_x={n_x}, n_y={n_y}, n_z={n_z})')

plt.show()
```
