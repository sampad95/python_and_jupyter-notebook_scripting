```python
toss_results = []
lst_rel_freq_head = []
n = 50000
for i in range(n):
    toss_results += [random.randint(0, 1)]
    head_result = toss_results.count(1)
    # print(head_result)
    rel_freq_head = head_result/len(toss_results)
    lst_rel_freq_head += [rel_freq_head]

# print(toss_results)
# print(toss_results.count(1))
# print(len(toss_results))
# print(len(lst_rel_freq_head))
# print(lst_rel_freq_head)

plt.plot(range(len(toss_results)), lst_rel_freq_head)
plt.axhline(y=0.5, color='gray', linestyle='--')
plt.xlabel('no. of toss')
plt.ylabel('relative frequency of getting head')
```
