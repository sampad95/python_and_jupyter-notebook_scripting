```python
fig = plt.figure(figsize = [4, 4], linewidth = 50)
Cin = []
Cout = []
for i in range(10000):
    def vector(x, y, z):
        return x, y, z
    
    r = vector(rnd.random(), rnd.random(), 0)
    rx, ry, rz = r
    

    if math.sqrt(rx**2 + ry**2 + rz**2) <= 1:
        Cin.append(r)
        plt.plot(rx,ry,".", color='tomato')
        
    else:
        Cout.append(r)
        plt.plot(rx,ry,'.', color='teal')

mpi = 4*len(Cin)/(len(Cin) + len(Cout))
print(mpi) 
```

```python
fig = plt.figure(figsize = [4, 4], linewidth = 50)
Cin = []
Cout = []
for i in range(10000):
    def vector(x, y, z):
        return x, y, z
    
    r = vector(rnd.uniform(-1,1), rnd.uniform(-1,1), 0)
    rx, ry, rz = r
    

    if math.sqrt(rx**2 + ry**2 + rz**2) <= 1:
        Cin.append(r)
        plt.plot(rx,ry,".", color='tomato')
        
    else:
        Cout.append(r)
        plt.plot(rx,ry,'.', color='teal')

mpi = 4*len(Cin)/(len(Cin) + len(Cout))
print(mpi) 
```
