```python
k_B = cons.Boltzmann
#T = eval(input('Enter the temperature in K'))
#T = 200
mu = 0*cons.e
E = np.linspace(-0.1, 0.1, 1000)*cons.e

def f(E, mu, T):
    return 1/(1+np.exp((E-mu)/(k_B*T)))

#print(f(E, mu, T))
#print(E-mu)

fig=plt.figure(figsize=(3,4),linewidth=50)
ax = fig.gca()
for axis in ['top','bottom','left','right']:
    ax.spines[axis].set_linewidth(3)
#plt.figure(figsize=(8,6),linewidth=50)  
#ax.grid(linestyle=':')
plt.grid(linestyle='--')


# evenly_spaced_interval = np.linspace(0, 1, len(E-mu))
# colors = [cm.rainbow(x) for x in evenly_spaced_interval]
#print(colors)

temps = list(range(50,375,25))

for T in np.array(temps):
    plt.plot(f(E, mu, T), (E-mu)/cons.e, lw =2, label = T)

plt.legend(bbox_to_anchor=(1.01, 1), loc='upper left')
plt.show()
```
