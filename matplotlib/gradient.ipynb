```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function and its gradient
def f(x, y):
    return x * np.exp(-(x**2 + y**2))

def grad_f(x, y):
    df_dx = (1 - 2 * x**2) * np.exp(-(x**2 + y**2))
    df_dy = -2 * x * y * np.exp(-(x**2 + y**2))
    return df_dx, df_dy

# Create a grid of points
x = np.linspace(-2, 2, 30)
y = np.linspace(-2, 2, 30)
X, Y = np.meshgrid(x, y)

# Compute the function values and gradients
Z = f(X, Y)
df_dx, df_dy = grad_f(X, Y)

# Create the pseudocolor plot
plt.figure(figsize=(10, 6))
plt.contourf(X, Y, Z, levels=45, cmap='viridis')
plt.colorbar(label='f(x, y)')

# Add the gradient vector field as arrows
plt.quiver(X, Y, df_dx, df_dy, color='k', scale=35, width=0.002, headwidth=4, headlength=3, headaxislength=3)

# Add labels and title
plt.xlabel('x')
plt.ylabel('y')
plt.title(r'Gradient of $f(x, y) = x e^{-(x^2 + y^2)}$')

plt.show()
```
